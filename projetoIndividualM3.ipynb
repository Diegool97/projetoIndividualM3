**Importação das Bibliotecas**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**Configuração das Visualizações**

```python
sns.set(style="whitegrid")
```

**Criação do DataFrame com os dados fornecidos**

```python
data = {
    "Dia": ["Segunda", "Terça", "Quarta", "Quinta", "Sexta", "Sábado", "Domingo"],
    "Horas Trabalhadas": [6, 7, 8, 6, 7, 5, 4],
    "Bugs Corrigidos": [3, 2, 1, 4, 3, 2, 1],
    "Tarefas Concluídas": [5, 4, 6, 4, 5, 3, 2]
}

df = pd.DataFrame(data)
```

**Exibição do DataFrame**

```python
df
```

**Calculo das Métricas**

```python
total_horas_trabalhadas = df["Horas Trabalhadas"].sum()
media_diaria_horas_trabalhadas = df["Horas Trabalhadas"].mean()
total_bugs_corrigidos = df["Bugs Corrigidos"].sum()
media_diaria_bugs_corrigidos = df["Bugs Corrigidos"].mean()
total_tarefas_concluidas = df["Tarefas Concluídas"].sum()
media_diaria_tarefas_concluidas = df["Tarefas Concluídas"].mean()
df["Produtividade Diária (Tarefas/Hora)"] = df["Tarefas Concluídas"] / df["Horas Trabalhadas"]
```

**Exibição das Métricas**

```python
metricas = {
    "Total de Horas Trabalhadas": total_horas_trabalhadas,
    "Média Diária de Horas Trabalhadas": media_diaria_horas_trabalhadas,
    "Total de Bugs Corrigidos": total_bugs_corrigidos,
    "Média Diária de Bugs Corrigidos": media_diaria_bugs_corrigidos,
    "Total de Tarefas Concluídas": total_tarefas_concluidas,
    "Média Diária de Tarefas Concluídas": media_diaria_tarefas_concluidas
}

metricas
```

**Total de Horas Trabalhadas e Média Diária**

```python
plt.figure(figsize=(10, 6))
sns.barplot(x="Dia", y="Horas Trabalhadas", data=df, palette="viridis")
plt.title("Horas Trabalhadas por Dia")
plt.xlabel("Dia")
plt.ylabel("Horas Trabalhadas")
plt.axhline(y=media_diaria_horas_trabalhadas, color='r', linestyle='--', label=f'Média Diária ({media_diaria_horas_trabalhadas:.2f})')
plt.legend()
plt.show()
```

**Total de Bugs Corrigidos e Média Diária**

```python
plt.figure(figsize=(10, 6))
sns.barplot(x="Dia", y="Bugs Corrigidos", data=df, palette="magma")
plt.title("Bugs Corrigidos por Dia")
plt.xlabel("Dia")
plt.ylabel("Bugs Corrigidos")
plt.axhline(y=media_diaria_bugs_corrigidos, color='r', linestyle='--', label=f'Média Diária ({media_diaria_bugs_corrigidos:.2f})')
plt.legend()
plt.show()
```

**Total de Tarefas Concluidas e Média Diária**

```python
plt.figure(figsize=(10, 6))
sns.barplot(x="Dia", y="Tarefas Concluídas", data=df, palette="coolwarm")
plt.title("Tarefas Concluídas por Dia")
plt.xlabel("Dia")
plt.ylabel("Tarefas Concluídas")
plt.axhline(y=media_diaria_tarefas_concluidas, color='r', linestyle='--', label=f'Média Diária ({media_diaria_tarefas_concluidas:.2f})')
plt.legend()
plt.show()
```

**Produtividade Diária**

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x="Dia", y="Produtividade Diária (Tarefas/Hora)", data=df, marker='o', color='b')
plt.title("Produtividade Diária (Tarefas Concluídas por Hora)")
plt.xlabel("Dia")
plt.ylabel("Produtividade Diária (Tarefas/Hora)")
plt.show()
```
